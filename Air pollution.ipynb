```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report as cr , confusion_matrix as cm
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
import numpy as np
```

```python
z = pd.read_csv(r"D:\SRM\Github codes\Datasets\data2\updated_pollution_dataset.csv")
z
```

```python
z["Air Quality"].value_counts()
```

```python
def change(s):
    if(s == "Good"):
        return(1)
    elif(s == "Moderate"):
        return(2)
    elif(s == "Poor"):
         return(3)
    else:
        return(0)
     
```

```python
z["Air Quality"] = z["Air Quality"].apply(change)
```

```python
b = z.copy()
for i in b:
    if(b[i].dtype == "object"):
        b.drop([i], axis = 1, inplace = True)
b
```

```python
b.corr()
```

```python
X = b.copy()
X.drop(["Air Quality"], axis = 1, inplace = True)
Y = b["Air Quality"]
```

```python
from sklearn.decomposition import PCA
```

```python
a = PCA()
x = a.fit_transform(X)
x.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, Y, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
y_train = np.array(y_train).reshape(-1, 1)
y_test = np.array(y_test).reshape(-1, 1)
n = RandomForestClassifier()
n.fit(x_train, y_train)
```

```python
y_predict_train = n.predict(x_train)
print(cr(y_true = y_train, y_pred = y_predict_train))
```

```python
y_predict_test = n.predict(x_test)
print(cr(y_true = y_test, y_pred = y_predict_test))
```

```python
cm(y_true = y_train, y_pred = y_predict_train)
```
